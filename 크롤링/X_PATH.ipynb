# 킥보드 자전거를 키워드로 가져오기

```python
import time

import requests
from lxml import html

keywords = ['킥보드', '자전거']



def crawl_news(keyword, page_num):
    url = 'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=' + keyword + '&sort=1&photo=0&field=0&pd=3&ds=2021.01.01&de=2021.04.30&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:dd,p:from20210101to20210430,a:all&start=' + str(
        page_num)
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}
    html_req = requests.get(url, headers=headers)
    tree = html.fromstring(html_req.text)
    bodies = tree.xpath('//ul[@class="list_news"]/li')
    results = []

    for body in bodies:

        news_title = body.xpath('.//a[@class="news_tit"]/@title')[0]

        try:
            news_url = body.xpath('.//a[@class="info"/@href')[0]
        except:
            news_url = ''
        news_title_clean = news_title.replace("\n", " ").replace("\t", " ").replace("\r", " ").strip()
        results.append(news_title_clean)

    return results

def fmain():
    for keyword in keywords:
        for i in range(1,4):
            print((i))
            page_num = (i-1) *10 +1
            results = crawl_news(keyword, page_num)
            print(results)
            time.sleep(6)
fmain()
```

```python
import time

import requests
from lxml import html

keywords = ['킥보드', '자전거']
#
# def make_file(keyword):
#     output_file_name = 'naver_news_' + keyword + "_" + time.strftime("%Y%m%d_%H%M%S") + '.txt'
#     output_file = open(output_file_name, "w", encoding="utf-8")
#     output_file.write("{}\t{}\t{}\t{}\n".format('페이지', '키워드', '제목', 'URL'))
#     output_file.close()
#     return output_file_name

#다른 폴더에 저장
path = 'data/'
def make_file(keyword):
    output_file_name = 'naver_news_' + keyword + "_" + time.strftime("%Y%m%d_%H%M%S") + '.txt'
    output_file = open(output_file_name, "w", encoding="utf-8")
    output_file.write("{}\t{}\t{}\t{}\n".format('페이지', '키워드', '제목', 'URL'))
    output_file.close()
    return output_file_name


def write_news(i, keyword, news_title_clean, news_url, output_file_name):
    print([i, keyword, news_title_clean, news_url])
    output_file = open(output_file_name, "a", encoding="utf-8")
    output_file.write("{}\t{}\t{}\t{}\n".format(i, keyword, news_title_clean, news_url))
    output_file.close()
    return

# def write_news(i, keyword, news_title_clean, news_url,output_file_name):
#     print([i,keyword,news_title_clean, news_url])
#     output_file = open(path+output_file_name, 'a', encoding="utf-8")               # 내용을 추가하려면 a
#     output_file.write("{}\t{}\t{}\t{}\n".format(i, keyword, news_title_clean, news_url))
#     output_file.close()
#     return


def crawl_news(keyword, i, output_file_name):
    page_num = (i - 1) * 10 + 1
    url = 'https://search.naver.com/search.naver' \
          '?where=news&sm=tab_pge' \
          '&query=' + keyword + '&sort=1&photo=0&field=0&pd=3' \
          '&ds=2021.01.01&de=2021.04.30&mynews=0&office_type=0' \
          '&office_section_code=0&news_office_checked=' \
          '&nso=so:dd,p:from20210101to20210430,a:all&start=' + str(page_num)
    print(url)
    headers ={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}
    html_req = requests.get(url, headers=headers)
    tree = html.fromstring(html_req.content)
    bodies = tree.xpath('//ul[@class="list_news"]/li')
    results = []

    for body in bodies:
        news_title = body.xpath('.//a[@class="news_tit"]/@title')[0]
        # news_url = body.xpath('.//a[@class="info"]/@href')[0]
        try:
            news_url = body.xpath('.//a[@class="info"]/@href')[0]
        except:
            news_url = ''
        news_title_clean = news_title.replace("\n", " ").replace("\t", " ").replace("\r", " ").strip()
        results.append([i, keyword, news_title_clean, news_url])
        write_news(i, keyword, news_title_clean, news_url, output_file_name)

    return results

def fmain():
    for keyword in keywords:
        output_file_name = make_file(keyword)
        for i in range(1, 4):
            print(i)
            results = crawl_news(keyword, i, output_file_name)
            print(results)
            time.sleep(6)


fmain()
```

```python
import time
import requests
from lxml import html

keyword = '킥보드'
input_file_name = 'naver_news_킥보드_20220328_020013.txt'

output_file_main_name = 'naver_news_main'+keyword+"_"+time.strftime("%y%m%d_%H%M%S")+'.txt'
output_file_main = open(output_file_main_name, "w", encoding='utf-8')
output_file_main.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format('번호', '키워드', '매체', '날짜', '제목', 'URL', '본문'))
output_file_main.close()
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36 Edg/95.0.1020.53'}

def get_list():
    input_file = open(input_file_name, "r", encoding="utf-8")
    input_text = input_file.read()
    lines = input_text.splitlines()
    lists = []

    for line in  lines:
        elms = line.strip().split("\t")
        news_title = elms[2]
        try:
            url = elms[3]
        except:
            url = ''
        lists.append([news_title, url])
    return lists[1:]  # 1부터 빼오는 이유는  0번 인덱스는 제목이기 때문

def write_news_main(count, news_media, news_date, news_title,news_url, news_article):
    print([count, news_media, news_date, news_title,news_url,news_article])
    output_file_main = open(output_file_main_name, 'a', encoding="utf-8")               # 내용을 추가하려면 a
    output_file_main.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(count, keyword, news_media, news_date, news_title, news_url, news_article))
    return


def crawl_news_main(count, news_title, news_url):
    html_req = requests.get(news_url, headers= headers)
    tree = html.fromstring(html_req.content)

    try:
        news_media = tree.xpath('//div[@class="press_logo"]/a/img/@title')[0]

    except:
        news_media = ''

    try:
        news_date = tree.xpath('//div[@class="sponsor"]/span[@class="t11"]/text()')[0]
    except:
        news_date = ''
    try:
        news_article = tree.xpath('//div[@id="articleBodyContents"]/descendant-orself::text()[not(ancestor::script)]')
    except:
        news_article = ''

    write_news_main(count,news_media,news_date,news_title,news_url,news_article)

    return

def fmain():
    lists = get_list()
    count = 1
    for list in lists:
        print(list)
        news_title = list[0]
        news_url = list[1]

        if len(news_url) == 0 :
            continue
        crawl_news_main(count,news_title,news_url)
        time.sleep(6)
        count += 1
fmain()
```

```python
import requests
from lxml import html
import time

# 파일 읽어오기
keyword = 'pann'
input_file_name = 'NATEpann_list_220328_020957.txt'

# 파일 저장하기  ( 열머리글 : 번호 글쓴이 날짜 조회수 추천수 댓글수 제목 URL 보문)
output_file_main_name = keyword + "_main_" + time.strftime("%y%m%d_%H%M%S") + '.txt'
output_file_main = open(output_file_main_name, "w", encoding="utf-8")
output_file_main.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format('번호', '글쓴이', '날짜', '조회수', '추천수', '댓글수', '제목', 'URL', '본문'))
output_file_main.close()

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:82.0) Gecko/20100101 Firefox/82.0'}


#
def get_list():
    input_file = open(input_file_name, "r", encoding="utf-8") # input 변수를 열어서 input_file에 저장,
    input_text = input_file.read()                            # text
    lines = input_text.splitlines()                           # line
    lists = []
    # print("lines:", lines)
    for line in lines[:]:
        elms = line.strip().split("\t")
        print(elms)
        page_num = elms[0]
        title = elms[1]
        try:
            url = elms[2]
        except:
            url = ''
        lists.append(  [page_num, title, url]    )      #   list = pagenum , title, rul   이렇게 출력됨!

    return lists[1:]    # list[0] = 열머리글이기 때문에 해당하는 열의 내용을 출력하려면 1번 인덱스부터~


# 위에서 저장하고자 하는 내용의 수와 맞게 변수 설정
def write_article_main(count, article_media, article_date, view_cnt, thumb_cnt, reply_cnt, article_title, article_url, article_article):
    print([count, article_media, article_date, view_cnt, thumb_cnt, reply_cnt, article_title, article_url, article_article])
    output_file_main = open(output_file_main_name, "a", encoding="utf-8")
    output_file_main.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(count, article_media, article_date, view_cnt, thumb_cnt, reply_cnt, article_title, article_url, article_article))
    return

#
def crawl_article_main(count, article_title, article_url):
    html_req = requests.get(article_url, headers=headers)
    tree = html.fromstring(html_req.content)

    try:
        article_media = tree.xpath('//a[@class="writer"]/text()')[0]
    except:
        article_media = ''
    try:
        article_date = tree.xpath('//span[@class="date"]/text()')[0]
    except:
        article_date = ''
    try:
        view_cnt = tree.xpath('//span[@class="count"]/text()')[0]
    except:
        view_cnt = 0
    try:
        thumb_cnt = tree.xpath('//div[@class="btnbox up"]/span[@class="count"]/text()')[0]
    except:
        thumb_cnt = 0
    try:
        reply_cnt = tree.xpath('//div[@class="cmt_tit"]/span[@class="num"]/strong/text()')[0]
    except:
        reply_cnt = 0
    try:
        article_article = tree.xpath('//div[@id="contentArea"]/descendant-or-self::text()[not(ancestor::script)]')
    except:
        article_article = ''
    article_article = ''.join(article_article)
    article_article = article_article.replace('\n', ' ').replace('\t', ' ').replace('\r', ' ').replace('\xa0', ' ').replace('\u200b', ' ').strip()

    write_article_main(count, article_media, article_date, view_cnt, thumb_cnt, reply_cnt, article_title, article_url, article_article)

    return



def fmain():
    lists = get_list()
    count = 1
    for list in lists[:]:
        print(list)
        article_title = list[1]
        article_url = list[2]
        if len(article_url) == 0:
            continue
        crawl_article_main(count, article_title, article_url)
        # time.sleep(6)
        count += 1


fmain()
```
