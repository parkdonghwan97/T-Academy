```python
df = pd.read_csv('data/cars.csv') 
df.head(3)
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
# units= 라벨의 갯수, input_dim = 특성데이터 갯수 
dense = Dense( units=1, input_dim=1 )   # w b 초기값
model = Sequential([dense])
model.compile(loss='mse',optimizer =Adam(0.1))
h = model.fit(x_data ,y_data, epochs=500)
```

```python
dense.get_weights() # w 와 b 둘다 구함. 3.0161774     -1.773635
```

```python
w,b = dense.get_weights()
```

```python
model.predict([10]) # 속도가 10일때 제동거리
```

```python
model.predict([10,14]) # 속도가 10, 14일 때 제동거리
```

```python
import torch
from torch.optim import Adam
```

```python
x_data = torch.FloatTensor([1,2,3,4,5])
y_data = torch.FloatTensor([2,4,6,8,10])
```

```python
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)
```

```python
optimizer = Adam( [w,b],lr=0.1)   
for i in range(2000):
    hx = w * x_data + b 
    cost = torch.mean( (hx -y_data)**2 )
    optimizer.zero_grad() # 기울기 초기화
    cost.backward() # cost값 미분
    optimizer.step() # w =  w-0.1 x w미분값, b= b-0.1 x b미분값
    print(i,'cost: ',cost.item())
```
