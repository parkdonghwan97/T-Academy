```python
# 다중 분류 위한 예제
df = pd.read_csv('tdata/softmax.txt', sep =' ' , header = None)
df
```

```python
x_data = df.iloc[:,0:2]
y_data = df.iloc[:,2:]
```

```python
x_data.shape , y_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
initX =tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[2,3] ) ) #[특성데이터의 갯수, 라벳의갯수]
b = tf.Variable( initX( shape=[3] ) ) #라벨의 갯수
```

```python
def compute_loss():
    z = tf.matmul( x, w) + b
#     hx = tf.nn.softmax( z )
    cost_i = tf.nn.softmax_cross_entropy_with_logits( labels=y, logits=z )
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
optimizer = Adam( 0.1 )
for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    print( i, 'cost:', compute_loss().numpy() )
```

# 다중분류 연습
- iris 데이터


```python

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split


from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris['data'],columns=iris['feature_names'])
df['species'] = iris['target']
df
```

```python
# 원핫 인코딩 
temp = pd.DataFrame(data= pd.get_dummies(df['species']).values)
```

```python
from tensorflow.keras.utils import to_categorical
df = pd.concat([df,temp],  axis=1)
del(df['species'])
```


# 1. [5.1, 3.5, 1.4,  0.2   ]   종분류 예측

## tesnor

```python
x_data = df.iloc[:,0:4]
y_data = df.iloc[:,4:]
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
initX =tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[4,3] ) ) #[특성데이터의 갯수, 라벳의갯수]
b = tf.Variable( initX( shape=[3] ) ) #라벨의 갯수
def compute_loss():
    z = tf.matmul( x, w) + b
#     hx = tf.nn.softmax( z )
    cost_i = tf.nn.softmax_cross_entropy_with_logits( labels=y, logits=z )
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
optimizer = Adam( 0.1 )
for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    print( i, 'cost:', compute_loss().numpy() )
```

```python
def hxfn( xdata):
    xd = tf.constant( xdata ,tf.float32)
    z = tf.matmul( xd, w) + b
    hx = tf.nn.softmax( z )
    return hx.numpy()
#     return hx.numpy().argmax(axis=1)
#     return z.numpy(), hx.numpy()
```

```python
#  해당 x 데이터 예측 확률
hxfn([[5.1, 3.5, 1.4, 0.2 ] ])
```

```python
# 해당 x데이터 예측 값
hxfn([[5.1, 3.5, 1.4, 0.2 ] ]).argmax(axis=1)
```

## keras

```python
dense = Dense( units=3, input_dim=4 ,activation="softmax") # w:2x3, b:3
model = Sequential( [dense])
model.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=1000)
```

```python
w,b = dense.get_weights()
w, b
```

```python
model.predict([[5.1, 3.5, 1.4, 0.2 ]])
```

```python
model.predict([[5.1, 3.5, 1.4, 0.2 ]]).argmax(axis=1)
```
