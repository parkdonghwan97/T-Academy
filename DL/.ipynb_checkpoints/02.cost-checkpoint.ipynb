````python
x_data = [1,2,3]
y_data = [1,2,3]
````

````python
plt.figure(figsize=(7,7))
plt.scatter(x_data,y_data)
plt.show()       
````

# 미분
```
상수 값을 미분하면 0   
y =  2   미분 -> 0     

y = x   미분 - > 1

y = 2x  미분 - > 2

y = x^2 미분 - > 2x

y = x^3 미분 - > 3x^2




w = 10            a 는 임의의 값
w = w - (런닝레이트(a) * 기울기)


ex) 

w = 10  -  0.01 X (w가 10인 경우의 기울기)

w = 4.5 - 0.01 X ( w가 4.5인 경우의 기울기)
...



````python
def cost( x, y, w ):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+ (hx-y[i])**2
    return c/len(x)
````

````python

for w in np.linspace(-3,5, 50):
    c = cost(x_data,y_data, w)
    print('w= {0} , cost = {1}'.format(w,c))
    plt.plot(w,c, 'mx')
plt.xlabel('w')
plt.ylabel('cost')
plt.show()
````
