````python
from numpy import array
from tensorflow.keras.models import Sequential    
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, GRU
````

````python
x = array( [[1,2,3], [2,3,4], [3,4,5], [4,5,6]] )
y = array([4,5,6,7])
print('x shape : ', x.shape) # (4,3)
print('y shape : ', y.shape) # (4,)
print(x)
````

````python
x = x.reshape( 4,3,1 ) # ( batch size, sequence length, hidden size)
x
````

````python
model = Sequential()
model.add( SimpleRNN(4, activation='relu', input_shape=(3,1) ) )
# model.add( Dense(units=32) )
model.add( Dense(units=1)  )
model.compile( loss='mse', optimizer='adam')
model.fit( x,y, epochs=1000)
````

# Many to Many 
----


````python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN,GRU,TimeDistributed
````

````python
X = list()
Y = list()
X = [x for x in range(5, 301, 5)]
Y = [y for y in range(20, 316, 5)]

X = np.array(X).reshape(20, 3, 1)
Y = np.array(Y).reshape(20, 3, 1)
````

````python
print(X)
print('='*30)
print(Y)
````

````python
X.shape,Y.shape
````

````python
X[0] # 5 10 15 
````

````python
Y[0] #  20, 25, 30  
````

TimeDistributed(Dense(1)  
-   cost ()       weight  RNN , LSTM 

````python
model = Sequential() 
model.add(LSTM(64, input_shape=(3,1),activation='relu',return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(loss = 'mse',optimizer = 'adam')
model.fit(X,Y,epochs=1000)
````

````python
arr = np.array([300,305,310]) 
model.predict(arr.reshape(1,3,1))
# 300 305 310  -> 316 320 325 
````
