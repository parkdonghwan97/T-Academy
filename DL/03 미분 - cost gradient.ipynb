## 합성함수의 미분

````python
# y = wx 
def cost(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w * x[i]                # 예측값
        c = c + (hx - y[i]) ** 2
    return c/len(x)

def gradient(x,y,w):
    c = 0
    for i in range(len(x)):
        hx = w * x[i]                # 예측값
        c = c+ (hx - y[i]) * x[i]
    return c/len(x)
````

````python
x_data = [1,2,3]
y_data =[1,2,3]
````

````python
w = 10
for epoch in range(200):
    c = cost(x_data,y_data,w)
    g = gradient(x_data,y_data,w)         # 미분한 값 : 기울기   
    w = w - 0.1 *g                       # 런닝레이트는 0.1     w = w - 0.1 * g
    print(epoch, 'cost=',c, 'w=',w)
    
    
````

# 넘파이를 이용한 cost 함수와 gradient 함수를 만들고 epoch 200번을 최적의 w


````python
x_data = np.array([1,2,3])
y_data = np.array([1,2,3])
````

````python
def cost(x,y,w):
    for i in range(len(x)):
        hx = w * x[i]         # 예측값  [1,2,3] * -1 =  [-1, -2, -3]
        c = sum( (hx-y)**2)   # [-1,-2,-3 - [1,2,3]] = [-2, -4, -6] ** 2  = [4, 16, 36]
    return c  / len(x)
````

````python
cost(x_data,y_data,-1)
````

````python
def gradient( x, y , w):
    c = np.sum( (w*x-y)*x*2  )
    return c/len(x)
````

````python
gradient(x_data,y_data,-1)
````

````python
w = 10
for epoch in range(200):
    c = cost( x_data, y_data, w)
    g = gradient( x_data,y_data, w)
    w = w-0.1*g
    print( epoch, "cost=", c, "w=", w)
print("w=", w)
````
