```python
def split_xy1(dataset, time_steps):
    x, y = list(), list()
    for i in range(len(dataset)):
        end_number = i + time_steps
        if end_number > len(dataset) -1:
            break
        tmp_x, tmp_y = dataset[i:end_number], dataset[end_number]
        x.append(tmp_x)
        y.append(tmp_y)
    return np.array(x), np.array(y)
```

```python
def make_dataset(data, label, window_size=20):
    feature_list = []
    label_list = []
    for i in range(len(data) - window_size):
        feature_list.append(np.array(data.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

## 3) time step 나누는 함수(function) 1 사용

```python
def split_xy1(dataset, time_steps):
    x, y = list(), list()  # x, y리스트 생성
    #  datset길이만큼 
    for i in range(len(dataset)):
        end_number = i + time_steps  # time step + 1
        
        if end_number > len(dataset) -1: 
            break          # 0~2까지 슬라이싱 (end_number = 3 부터 시작)
        tmp_x, tmp_y = dataset[i:end_number], dataset[end_number]
        x.append(tmp_x)
        y.append(tmp_y)
    return np.array(x), np.array(y)
```

```python
# df['passengers'] 를 넣든, df['passengers'].values 를 넣든 결과는 같다.

# x, y = split_xy1( df['passengers'].values, 3)
x, y = split_xy1( xN, 3)
print(x)  ## x 데이터는 reshape 이 필요: x.reshape(-1, 3, 1) 
print('='*40)
print(y)
```
