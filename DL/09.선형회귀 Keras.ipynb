```python
df = pd.read_csv('data/cars.csv',index_col='Unnamed: 0') 
df.head(3)
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
# units= 라벨의 갯수, input_dim = 특성데이터 갯수 
dense = Dense( units=1, input_dim=1 )   # w b 초기값
model = Sequential([dense])
model.compile(loss='mse',optimizer =Adam(0.1))
h = model.fit(x_data ,y_data, epochs=500)
```

```python
dense.get_weights() # w 와 b 둘다 구함. 3.0161774     -1.773635
```

```python
w,b = dense.get_weights()
```

```python
model.predict([10]) # 속도가 10일때 제동거리
```
