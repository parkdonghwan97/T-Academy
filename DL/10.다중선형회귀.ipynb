# 텐서플로우

```python
x1_data = [1,0,3,4,5]  #공부한시간
x2_data = [0,1,2,3,4] #출석일수
y_data = [1,2,3,4,5] #점수
```

```python
x1 = tf.constant( x1_data , tf.float32)
x2 = tf.constant( x2_data , tf.float32)
y = tf.constant( y_data , tf.float32)
```

```python
tf.random.uniform([1])
```

```python
w1 = tf.Variable(tf.random.uniform([1])) # x1 의 가중치
w2 = tf.Variable(tf.random.uniform([1])) # x2 의 가중치
b = tf.Variable(tf.random.uniform([1])) 
```

```python
# cost 구하는 함수
def compute_loss():
    hx = w1*x1 + w2*x2 + b
    cost = tf.reduce_mean( (hx - y)**2 )
    return cost
```

```python
# 
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list=[w1,w2,b] ) # 미분
    print(i,'cost: ',compute_loss().numpy())
```

#  MatMul ( 행렬곱 이용)

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
```

```python
# Matmul 사요해야하지때문에 q1 q2 mid 를 n행 3열 매트릭스로

x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
```

```python
x_data.shape, y_data.shape
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
''' 
3행 1열     
q1
w2
mid
'''
w = tf.Variable(tf.random.uniform(  [3,1] )   ) # 특성데이터갯수, label개수
b = tf.Variable(tf.random.uniform([1]) )   # [라벨개수]
```

```python
def compute_loss():
    hx = tf.matmul(x,w) + b
    cost = tf.reduce_mean( (hx-y)**2  )
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss,var_list=[w,b] )
    print(i,'cost: ',compute_loss().numpy())
```

# 케라스

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
```

```python
x_data = df.iloc[:,:-1].values #특성데이터 추출
y_data = df.iloc[:, [-1]].values
```

```python
dense = Dense(units = 1, input_dim =3 ) # w : 3x1   b : 1   초기값
model = Sequential([dense])
model.compile(loss='mse',optimizer=Adam(0.1))
# n행 3열
h = model.fit(x_data, y_data, epochs=500) # x값을 행렬로 주어야함
```
