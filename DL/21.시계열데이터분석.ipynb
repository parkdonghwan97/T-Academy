# 이동평균
- 전체 데이터 집합의 여러 하위 집합에 대한 일련의 평균을 만들어 데이터 요소를 분석

```python
data1 = [[1,2],[3,4],[5,8],[7,6],[9,12],[11,10],[13,16],[15,20]]
df = pd.DataFrame( data1, columns = ['A','B'] )
df.index=['2019-01-01','2019-02-01','2019-03-01','2019-04-01',
          '2019-05-01','2019-06-01','2019-07-01','2019-08-01']
df
```

```python
df.rolling(window=2).mean()  # sum, min, max   


# (1,2)의평균, (2,4)의 평균
# (3,5)의평균, (4,8)의 평균
#  ....

# (13,15)의평균,(16,20)의평균

```

```python
df
```

```python
df.rolling(window=3).mean()  # sum, min, max   

# (1,2),4의평균, (2,4,8)의 평균
# (3,5,7)의평균, (4,8,6)의 평균

# ...

# (11,13,15)의평균,(10,16,20)의평균

```

```python
plt.figure(figsize=(15,15))
plt.plot(df.index,df['B'])
plt.plot(df.index,df.rolling(window=2).mean()['B'] )
plt.xticks(rotation=45)
```

## 연도별 평균 
- 시계열데이터의 통계(집계)

### 년도별 평균

```python
sdf.groupby('year').mean()
```

```python
sdf.groupby(sdf.index.year).mean()
```

### 월별 평균

```python
sdf.groupby(sdf.index.month).mean()
```

### (년도-월)별 평균

```python
sdf.groupby( [sdf.index.year, sdf.index.month]  ).mean()
```

# resample : 시계열 데이터 전용 집계함수





DatetimeIndexResampler : sum, mean,max,min

```python
sdf.resample(rule='A').mean() # A : 년도별
```

```python
sdf.resample(rule='M').mean() # M : 월별 집계
```

```python
sdf.resample(rule='W').mean() # W : 주간별 집계결과
```

```python
sdf.resample(rule='Q').mean() # Q : 분기별 집계
```
