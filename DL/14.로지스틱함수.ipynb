````python
def compute_loss():
#     hx = tf.matmul(x,w) + b
    z = tf.matmul(x,w) + b
    hx = tf.nn.sigmoid(z)  # (뉴럴 네트워크)
    

#     cost = tf.reduce_mean((hx-y)**2)

    # 라벨 = 정답(y)   logits= 시그모이드 (z)
    # 미분이 안되기 때문에 로그 씌워서 
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels=y, logits=z)
    cost = tf.reduce_mean(cost_i)
    return cost
    
````

````python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    print(i, 'cost:', compute_loss().numpy() )
````

````python

def hxfn(xdata):
    xd = tf.constant(xdata,tf.float32)
    z = tf.matmul(xd,w ) + b
    hx = tf.nn.sigmoid(z)
#     return z.numpy(),hx.numpy()
    return (hx.numpy() > 0.5) + 0    # 1 일경우 합, 0일경우 불합.
````

# 파이프라인 사용
- 인디언 당뇨병 

````python
import pandas as pd
df = pd.read_csv('ddata/pima-indians-diabetes.data.csv',encoding='euc-kr')
# 필요컬럼 
for c in ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']:
    idx = np.where(df[c]==0)[0]
    df.loc[idx,c] = df[c].mean()
df.head(3)
````

````python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
````

````python
# 스케일링
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_data)

x = tf.constant(x_scaled, tf.float32)
y = tf.constant(y_data, tf.float32)
````

````python
def createModel():
    dense = Dense( units=1, input_dim=8 ,activation="sigmoid") # w:2x1, b:1
    model = Sequential( [dense])
    model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )

    return model
#h = model.fit( x_data, y_data, epochs=1000)
````

````python
dense = Dense( units=1, input_dim=8 ,activation="sigmoid") # w:2x1, b:1
model = Sequential( [dense])
model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_scaled,y, epochs=1000)
````

# 파이프라인 - 스케일링한 데이터 적용

````python
kcf = KerasClassifier(build_fn = createModel, epochs=1000 ,verbose=1)  # 케라스 분류모델
pmodel = make_pipeline(MinMaxScaler(),kcf)
pmodel.fit(x_data,y_data)
````
